```python
#importing csv
import pandas as pd
import re

df = pd.read_csv("/Users/jodieengland/Documents/work/cvs-and-job-apps/Jobs/Rockstar-Games/gaming-review-analysis/data/steam_reviews.csv")
```

```python
#defining the clean_text function
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)     #remove links
    text = re.sub(r"[^a-z\s]", "", text)    #remove punctuation
    return text
```

```python
#running the function
df['clean_review'] = df['review'].apply(clean_text)
```

```python
#removing empty reviews
df = df[df['clean_review'].str.strip() != ""]
```

```python
#export to csv
df.to_csv("/Users/jodieengland/Documents/work/cvs-and-job-apps/Jobs/Rockstar-Games/gaming-review-analysis/data/clean_reviews.csv", index=False)
```
